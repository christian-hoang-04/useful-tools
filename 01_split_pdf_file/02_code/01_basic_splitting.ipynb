```python
# Step 1: Import PyPDF2 and Prepare the Input File 
from pypdf import PdfReader, PdfWriter 
```

```python
# Split each page into new PDF files

start_page = 1 
end_page = 3 
input_pdf_path = r"D:\Github\useful-tools\01_split_pdf_file\01_sample\01_inputs\input.pdf"
output_folder_path = r"D:\Github\useful-tools\01_split_pdf_file\01_sample\02_outputs"
file_name = "output_page"

# Step 2: Open the PDF file 
with open(input_pdf_path, "rb") as input_file: 
    reader = PdfReader(input_file) # Create a PdfReader object

    # Step 3: Get the Number of Pages 
    num_pages = len(reader.pages)
    print(f"Input PDF has {num_pages} pages.")

    # Step 4: Split and Save Individual Pages 
    for page_num in range(start_page-1, end_page): 
        # Create a new PDF writer for each page 
        writer = PdfWriter()

        # Add the current page to the writer 
        page = reader.pages[page_num]
        writer.add_page(page)

        # Define output filename (e.g., "output_page_1.pdf" for page 0)
        output_filename = rf"{output_folder_path}\{file_name}_{page_num + 1}.pdf" # + 1 to make pages 1-based 
        
        # Write the new PDF to a file 
        with open(output_filename, "wb") as output_file: # 'wb' = write binary mode 
            writer.write(output_file)
        
        print(f"Saved: {output_filename}")
```
